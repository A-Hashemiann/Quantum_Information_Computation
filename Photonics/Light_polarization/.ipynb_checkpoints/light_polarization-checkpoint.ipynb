```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
amplitude = 1 
wavelength = 500e-9 
frequency = 3e8 / wavelength  # (Hz)
angular_frequency = 2 * np.pi * frequency  
k = 2 * np.pi / wavelength  
time = 0  # (seconds)

# generate positions along the propagation axis
x = np.linspace(0, 5 * wavelength, 1000)  # positions along the propagation axis (x)

# electric field (E) oscillates in the y-direction
E = amplitude * np.cos(k * x - angular_frequency * time)

# magnetic field (B) oscillates in the z-direction and is scaled for distinction
B = 0.3 * amplitude * np.cos(k * x - angular_frequency * time)  # Scaling to 30% of E

# plotting
plt.figure(figsize=(10, 6))
plt.plot(x, E, label="Electric Field (E)", color="blue", linewidth=2)
plt.plot(x, B, label="Magnetic Field (B)", color="red", linewidth=2)


plt.title("Electric and Magnetic Fields of an Electromagnetic Wave")
plt.xlabel("Position (x) [m]")
plt.ylabel("Field Amplitude")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")  # zero line for reference
plt.legend()
plt.grid(True)

# show the plot
plt.show()
```

```python

# create a 3D plot
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# electric field (E) in the y-direction
ax.plot(x, E, zs=0, zdir='z', label="Electric Field (E)", color='blue', linewidth=2)

# magnetic field (B) in the z-direction
ax.plot(x, 0 * x, zs=B, zdir='y', label="Magnetic Field (B)", color='red', linewidth=2)

# directions
ax.quiver(0, 0, 0, 1e-6, 0, 0, color='black', label="Propagation Direction (x)", arrow_length_ratio=0.2)

# visualization
ax.view_init(elev=20, azim=-60)

#labels
ax.set_xlabel("Propagation Direction (x)")
ax.set_ylabel("Electric Field (y)")
ax.set_zlabel("Magnetic Field (z)")


ax.set_title("3D Visualization of Electric and Magnetic Fields")
ax.legend()

# plot
plt.show()
```

```python


B_scale = 1  # ehanced scaling factor for the magnetic field (B)

# Generate positions along the propagation axis
x = np.linspace(0, 5 * wavelength, 100)  # x positions for propagation

# Electric field (E) oscillates in the y-direction
E = amplitude * np.cos(k * x - angular_frequency * time)

# Magnetic field (B) oscillates in the z-direction with enhanced scaling
B = B_scale * amplitude * np.cos(k * x - angular_frequency * time)




# Create a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection="3d")

# Plot the electric field (E) as blue arrows
for i in range(len(x)):
    ax.quiver(x[i], 0, 0, 0, E[i], 0, color="blue", label="Electric Field (E)" if i == 0 else "")

# Plot the magnetic field (B) as red arrows
for i in range(len(x)):
    ax.quiver(x[i], 0, 0, 0, 0, B[i], color="red", label="Magnetic Field (B)" if i == 0 else "")

# Add the propagation direction (x-axis) as a green line
ax.plot(x, np.zeros_like(x), np.zeros_like(x), color="green", label="Propagation Direction (k)")

# Set the axis limits to emphasize both fields
ax.set_xlim([0, 5 * wavelength])
ax.set_ylim([-1.5 * amplitude, 1.5 * amplitude])  # For E field range
ax.set_zlim([-1.5 * B_scale * amplitude, 1.5 * B_scale * amplitude])  # Enhanced for B field range

# Set labels and title
ax.set_xlabel("Propagation Direction (x)")
ax.set_ylabel("Electric Field Direction (y)")
ax.set_zlabel("Magnetic Field Direction (z)")
ax.set_title("3D Visualization of Electromagnetic Wave: E  B  k")
ax.legend()

# Adjust the aspect ratio for better clarity
ax.set_box_aspect([5, 1.5, 1.5])  # Proportional scaling of x, y, z

# Show the plot
plt.tight_layout()  # Adjust layout to fit everything properly
plt.show()
```
